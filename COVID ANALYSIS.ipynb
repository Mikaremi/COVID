```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
global_data = pd.read_csv(r'time_series_covid19_confirmed_global.csv')

global_data
```

Some countries have province/states while others dont.

I am going to aggreage the sum of all the countries    

```python
global_data_summed = global_data.groupby('Country/Region').sum()
```

```python
global_data_summed
```

I'll remove the Lat and Long columns because they are of no use since the dataframe has been summed

```python
df = global_data_summed.drop(['Lat', 'Long'], axis = 1).reset_index()
```

```python
df
```

I am going to  list out all the countries

```python
df['Country/Region'].unique()
```

```python
countries_to_plot = ['US', 'China', 'United Kingdom', 'Germany', 'Kenya']

# US date formatt --> MM/DD/YYYY
start_date = '01/22/2020'
end_date = '04/03/2020'
```

```python
#plotting

import datetime

def get_index_from_date(date_string):
    FIRST_DATE = datetime.datetime(2020, 1, 22) # Don't modify this value
    
    date_list = [int(x) for x in date_string.split('/')]
    date_object = datetime.datetime(date_list[-1], date_list[0], date_list[1])
    
    return (date_object-FIRST_DATE).days

dates = list(df)[1:]

start_index = get_index_from_date(start_date)
end_index = get_index_from_date(end_date)

x_values = dates[start_index:end_index+1]

for country in countries_to_plot:
    country_df = df[df['Country/Region'] == country]
    y_values = []
    for col in x_values:
        try:
            # Convert the value to an integer if it's a valid number
            y_value = int(country_df[col].iloc[0])
        except ValueError:
            # Handle cases where the value cannot be converted
            y_value = 0  # or use np.nan or any default value
        y_values.append(y_value)
    plt.plot(x_values, y_values, label=country)

    
plt.title('Confirmed Cases Over Time')
skip = max(len(x_values)//5, 1) # Helps ensure we don't add too many date tick marks
plt.xticks(x_values[::skip])
plt.xlabel("Date (MM/DD/YY)")
plt.ylabel("Confirmed Cases")
plt.legend()
plt.show()
```
